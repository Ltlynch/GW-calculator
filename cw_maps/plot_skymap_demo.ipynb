```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp
import numpy as np
import plot

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
nside = 4 #corresponds to 192 pixels
```

```python
skymap = np.loadtxt('uls_smoothed_10nHz.txt')
```

```python
ax = plt.subplot(111, projection='astro mollweide')
ax.grid()
plot.outline_text(ax)
plot.healpix_heatmap(np.log10(skymap[:,1]), cmap='viridis_r')

plt.colorbar(orientation='horizontal')
plt.grid(linestyle='dotted', color='k')
plt.suptitle('95% upper limit', y=0.01)

plt.tight_layout()
```
